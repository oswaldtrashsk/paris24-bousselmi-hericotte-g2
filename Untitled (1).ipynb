```python
import yfinance as yf
import pandas as pd
import datetime
from datetime import timedelta
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# 
def getting_gold_prices():
    gold = yf.Ticker("GC=F")
    gold_prices = gold.history(period="1d", interval="1m")
    
    # but then when i tried to run it, it didn't work because we are sunday, so there's no trading happening. the following code i made to find 
    # the last available data, meaning  it will find the data from friday on weekends. 
    # on stack overflow i found the index.max() part is used to find the most recent data available, and date allows to get
    # the date to use it as a reference and not a data
    if last_prices.empty:
        last_date = gold.history(period="5d", interval="1m").index.max().date() 
        #but this is for pandas' datetime, we need to adapt it to the standard library so yfinance can use it
        last_end_date = datetime(now.year, now.month, last_date) + timedelta(days=1, seconds=-1)
        last_start_date = last_end_date - timedelta(days = 1, seconds = -1)
        last_prices = gold.history(interval = "1m", start = last_start_date, end = last_end_date)

    return last_prices

# Calculate moving average
def calculate_moving_average(data, window=50):
    return data.rolling(window=window).mean()

# Plotting the data
def plot_gold_prices(prices):
    plt.figure(figsize=(10, 5))
    plt.plot(prices.index, prices, label='Gold Price', color='gold')
    plt.title(f'Gold Price - Data from last day available')
    plt.xlabel('Time')
    plt.ylabel('Price (USD)')
    plt.legend()
    plt.grid(True)
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
    plt.gcf().autofmt_xdate()
    plt.show()

# Analyze and give advice
def analyze_and_advice(prices):
    moving_average = calculate_moving_average(prices)
    current_price = prices.iloc[-1]
    advice = "Hold"  # Default advice
    if current_price > moving_average.iloc[-1]:
        advice = "Consider buying; trend may be upward."
    elif current_price < moving_average.iloc[-1]:
        advice = "Consider selling; trend may be downward."
    return advice

# Main function to run the program
def main():
    prices = fetch_gold_data()
    if prices.empty:
        print("No trading data available.")
        return
    plot_gold_prices(prices)
    advice = analyze_and_advice(prices)
    print("Current Time:", datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
    print("Trading Advice:", advice)

if __name__ == "__main__":
    main()
```

```python
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Fetch historical prices for gold, silver, and the S&P 500
gold_prices = yf.Ticker('GC=F').history(period="5y", interval="1d")['Close']
silver_prices = yf.Ticker('SI=F').history(period="5y", interval="1d")['Close']
sp500_prices = yf.Ticker('^GSPC').history(period="5y", interval="1d")['Close']

# Compute the covariance matrix
cov = pd.DataFrame({
    'Gold-Silver': gold_prices,
    'Gold-SP500': gold_prices,
    'Silver-SP500': silver_prices
}).cov()

# Compute the correlation matrix
corr = cov / np.sqrt(np.outer(cov.to_numpy().diagonal(), cov.to_numpy().diagonal()))

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(10, 5))
sns.heatmap(corr, annot=True)
corr.style.background_gradient(cmap="coolwarm")
```

```python
import yfinance as yf
import pandas as pd
import datetime
from datetime import timedelta
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

#we are getting the tickers for gold, the us dollar index, silver, and the SP 500 for comparison
tickers = ['GC=F', 'SI=F','DX-Y.NYB','^GSPC']

# and we can then  find the closing prices of the four indexes every week for a range of time, here i used period_ so i can change if needed
period_ = ('5y')

gold = yf.Ticker(tickers[0]).history(period = period_, interval = '1d')['Close']
silver = yf.Ticker(tickers[1]).history(period = period_, interval = '1d')['Close']
dollar = yf.Ticker(tickers[2]).history(period =period_, interval = '1d')['Close']
sp = yf.Ticker(tickers[3]).history(period = period_, interval = '1d')['Close']

# now,  we need a way to group all data in one table and rename the columns: 
data_raw = pd.concat([gold, silver, dollar, sp], axis=1)
data_raw.columns = ['Gold', 'Silver', 'US Dollar', 'S&P 500']

# then i found one problem being that many data were missing so i used the .dropna function to erase those missing data
clean_data = data_raw.dropna()

#and we can check to see if any data is missing a second time :
print(clean_data.isnull().isnull().sum())
```
